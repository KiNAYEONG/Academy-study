```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 가상의 데이터를 만들어 선 그래프
x = np.arange(1,6)
y = np.array([9,4,9,9,1])

# 선 그리는 방법
# plt.plot(x축, y축 데이터)
plt.plot(x,y)

plt.show()
```

```python
# x축이 늘어나는 숫자 생략 가능 y축만 대입
plt.plot(y)
plt.show()
```

```python
# 그래프 크기 조정
plt.figure(figsize=(3,2))
plt.plot(x,y, c = 'g') #c는 color의 약자

plt.show()
```

```python
# 한 셀에 여러 그래프 그리기
x1 = np.arange(1,6)
y1 = np.array([2,4,1,1,5])
# plt.plot을 이어서 쓸 경우 그래프가 겹쳐서 출력
plt.plot(x,y, ls = '-', c = 'b', marker = 'D', mec = 'r', mfc='y')
plt.plot(x1,y1, ls = '--', c = 'm', marker = '+', mfc = 'c')
plt.show()
```

```python
# plt.subplot(행, 열, 몇번째)
plt.subplot(2,2,1)
plt.plot(x,y, ls = '-', c = 'b', marker = 'D', mec = 'r', mfc='y')
plt.subplot(2,2,2)
plt.plot(x1,y1, ls = '--', c = 'm', marker = '+', mfc = 'c')
plt.show()
```

```python
# 차트의 이름 
# 격자눈금
plt.plot(x,y)
plt.title('Test')
plt.grid()
```

```python
# legend
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")
plt.legend() #레전드 사용 위해선 plot label 이름 지정, 코드는 위에서 아래로 내려오기 때문에, plot 다음에 써야 실행됨.
# loc를 작성하여 legend 위치 수정 가능 plt.legend(loc=10)
# shift + tab 눌러서 loc code 확인하기
```

```python
# xlim, ylim - 범위 제한
#(시작, 끝) == (1,10) 1~10
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")
plt.xlim()
```
