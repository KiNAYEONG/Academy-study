```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 가상의 데이터를 만들어 선 그래프
x = np.arange(1,6)
y = np.array([9,4,9,9,1])

# 선 그리는 방법
# plt.plot(x축, y축 데이터)
plt.plot(x,y)

plt.show()
```

```python
# x축이 늘어나는 숫자 생략 가능 y축만 대입
plt.plot(y)
plt.show()
```

```python
# 그래프 크기 조정
plt.figure(figsize=(3,2))
plt.plot(x,y, c = 'g') #c는 color의 약자

plt.show()
```

```python
# 한 셀에 여러 그래프 그리기
x1 = np.arange(1,6)
y1 = np.array([2,4,1,1,5])
# plt.plot을 이어서 쓸 경우 그래프가 겹쳐서 출력
plt.plot(x,y, ls = '-', c = 'b', marker = 'D', mec = 'r', mfc='y')
plt.plot(x1,y1, ls = '--', c = 'm', marker = '+', mfc = 'c')
plt.show()
```

```python
# plt.subplot(행, 열, 몇번째)
plt.subplot(2,2,1)
plt.plot(x,y, ls = '-', c = 'b', marker = 'D', mec = 'r', mfc='y')
plt.subplot(2,2,2)
plt.plot(x1,y1, ls = '--', c = 'm', marker = '+', mfc = 'c')
plt.show()
```

```python
# 차트의 이름 
# 격자눈금
plt.plot(x,y)
plt.title('Test')
plt.grid()
```

```python
# legend
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")
plt.legend() #레전드 사용 위해선 plot label 이름 지정, 코드는 위에서 아래로 내려오기 때문에, plot 다음에 써야 실행됨.
# loc를 작성하여 legend 위치 수정 가능 plt.legend(loc=10)
# shift + tab 눌러서 loc code 확인하기
```

```python
# xlim, ylim - 범위 제한
#(시작, 끝) == (1,10) 1~10
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")
plt.xlim(1,5)
plt.ylim(1,5)
```

```python
#한글 폰트 깨질 경우
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
# xlabel, ylabel
# ("타이틀")
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")
plt.xlabel("X축")
plt.ylabel("Y축", rotation = 0, loc = 'bottom' )
# x , y축의 각도 수정을 희망시 rotation = 각도
plt.show()
```

```python
# xticks, yticks
plt.plot(x,y, label = 'test1')
plt.plot(x1,y1, label = "test2")

# 매개변수 list, array
plt.xticks(range(1,6))
plt.show()
```

```python
data = pd.read_csv('data/장래_인구변동_KOSIS.csv', encoding = 'euc-kr', index_col = "인구변동요인별")
data
```

```python
display(data.columns)

data_x = data.columns
```

```python
#y축을 출생아수
data_y = data.iloc[1]
```

```python
#y축을 사망자수
data_y1 = data.iloc[2]
```

```python
plt.plot(data_x, data_y, marker = "*" , ms = 10, label='출생아수(천명')
plt.plot(data_x, data_y1, marker = "o" , ms = 10, label='사망자수(천명')
plt.title('국가통계 요인별 장래 인구 동향')
plt.xlabel('연도')
plt.ylabel('사람수')
plt.yticks(range(150, 750, 50))
plt.grid()
plt.legend() 
plt.show()
```
