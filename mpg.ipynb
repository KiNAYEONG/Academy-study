```python
# np.where 사용 조건에 맞게 파생변수
# where(조건식, True, False)
mpg_copy['test'] = np.where(mpg_copy['평균'] >= 25, 'pass', 'fail')
mpg_copy
```

```python
# 복잡한 수식 사용
# 25점 이상 A등급 20점 B등급 15점 C등급 나머지 F등급
# .apply(함수명, axis = 1)
# ㄴ 데이터프레임명.apply(함수) --> 함수에게 매개변수로 데이터 프레임이 넘어감
mpg_copy['grade']=mpg_copy.apply(makeGrade, axis = 1)
mpg_copy
```

```python
def makeGrade(row):
    mean = row['평균']
    if mean >= 25 :
        return "A등급"
    elif mean >=20:
        return "B등급"
    elif mean >=15:
        return "C등급"
    else:
        return "F등급"
        
```

### groupby
- 데이터를 그룹별로 묶어서 집계(합, 평균 중앙값)
- 기준을 잡고 데이터를 묶기

```python
# mpg_copy 데이터를 도시를 기준으로 묶기
mpg_copy.groupby('도시')
# 그룹화 된 직접적으로 보기 힘드나 집계 함수들을 사용하여 출력가능

mpg_copy.groupby('도시').mean(True)
#도시연비가 9인것들끼리 묶임. 그것들의 평균을 구한 것
```

```python
# 제조사끼리 groupby
# 반드시 집계함수까지 같이 써야 출력이 된다!
mpg_copy.groupby('제조사').mean(True)
```

## groupby 집계 함수
- mean()
- sum() : 그룹별 합계
- count() : 개수 확인
- median() : 중앙값 확인
- min() : 최소값
- max() : 최대값
- agg() : 다중 집계 작업(집계함수를 다중으로 적용 가능)

# agg() : 다중 집계 함수
- 다중 집계 뿐만 아니라 데이터 프레임 생성하여 출력 가능

```python
# 1개 집계 함수 사용 agg(컬럼명(변수명) = (컬럼명(실제 있는 컬럼명), "집계함수"))  
## << 컬럼명 쓸 때 따옴표 안쓰고 가급적 영문 작성 추천
mpg_copy.groupby("제조사").agg(도시평균 = ('도시','mean'))
```

```python
# 1개의 컬럼에서 2개 이상의 집계함수
mpg_copy.groupby('제조사').agg({'도시' : ['mean', 'count']})
```

```python
#2개의 컬럼 - 1개 컬럼 출력 방식에서 ',' 작성 후 이어쓰기
mpg_copy.groupby('제조사').agg(도시평균 = ('도시', 'mean'), 고속도로평균 = ('고속도로', 'mean'))
```

```python
mpg_copy.head(1)
```

```python
# 제조사별 등급별 차종 개수
# 제조사, 등급으로 묶기
mpg_copy.groupby(['제조사','grade']).agg(cnt = ('grade','count'))
```

```python
# 카테고리 별 도시 연비 평균
# 카테고리별로 묶고 도시 연비의 평균
mpg_copy.groupby('category').agg(city = ('도시', 'mean'))
mpg_copy.groupby('category').agg(city = ('도시', 'mean')).sort_values(by = 'city', ascending = False)
```

## 어떤 회사가 compact 차종을 가장 많이 생산

```python
# 1. xs 함수 - 다중 인덱싱 되어있는 데이터 프레임에서 특정 행 혹은 특정 열을 선택

# 1.1 제조사, category를 묶기
mpg_copy.groupby(['제조사','category'])
# 1.2 category별 개수 확인
mpg_copy.groupby(['제조사','category']).agg(cnt = ('category', 'count'))

# 1.3 xs 함수 .xs(선택할 값, level = 컬럼명)
mpg_copy.groupby(['제조사','category']).agg(cnt = ('category', 'count')).xs('compact', level = 'category')
```

```python
#2. 불리언 인덱싱 후 다중집계 함수
#2.1 category == compact
mpg_copy.query("category == 'compact'")
#2.2 제조사만 묶기
mpg_copy.query("category == 'compact'").groupby('제조사')
#2.3 category 개수 세기
mpg_copy.query("category == 'compact'").groupby('제조사').agg(cnt=('category','count'))
```

```python
# 3. 시리즈로 가지고오기
mpg_copy.query('category == "compact"')['제조사'].value_counts()
```
