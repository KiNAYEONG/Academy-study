```python
#   
# 25  A 20 B 15 C  F
# .apply(, axis = 1)
#  .apply() -->     
mpg_copy['grade']=mpg_copy.apply(makeGrade, axis = 1)
mpg_copy
```

```python
def makeGrade(row):
    mean = row['']
    if mean >= 25 :
        return "A"
    elif mean >=20:
        return "B"
    elif mean >=15:
        return "C"
    else:
        return "F"
        
```

```python
# 1. xs  -           

# 1.1 , category 
mpg_copy.groupby(['','category'])
# 1.2 category  
mpg_copy.groupby(['','category']).agg(cnt = ('category', 'count'))

# 1.3 xs  .xs( , level = )
mpg_copy.groupby(['','category']).agg(cnt = ('category', 'count')).xs('compact', level = 'category')
```
