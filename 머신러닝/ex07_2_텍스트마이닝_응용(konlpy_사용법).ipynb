```python
# 형태소분석기 도구 불러오기
from konlpy.tag import Okt,Kkma,Mecab
```

```python
# 객체생성
okt = Okt()       #트위터 기반으로 시작된 형태소 분석기 -> 비교적 신조어에 강하다.
kkma = Kkma()     #서울대학교에서 생성한 분석기, 속도가 조금 느림, 품사 태깅이 디테일한편
mecab = Mecab()   #처리속도가 빠른 것이 장점
```

```python
# 각 형태소 분석기별 형태소 분리 확인
print(okt.morphs('아버지가방에들어가신다.'))
print(okt.morphs('아버지 가방에 들어가신다.'))
```

```python
import emoji
```

```python
text_rm_emoji = [emoji.replace_emoji(s) for s in data['리뷰']]
```

```python
text_rm_emoji
```

```python
# 형태소 분석기 사용하기
from tqdm import tqdm as tq
```

```python
text_morphs = [mecab.morphs(s) for s in tq(text_rm_emoji)]
```

```python
# 정규표현식을 활용한 클렌징
import re
pattern_text = re.compile('[!?,.:;0-9a-zA-Zㅋㅎㄷㅇㅠ~]+')
```

```python
# 클렌징된 단어토큰들을 담을 비어있는 리스트 생성
clean_morphs = []
for s in text_morphs: # 형태소로 분리된 전체데이터 반복
  temp = [] # 한문장을 담을 리스트 생성
  for w in s:
    if pattern_text.search(w): # 단어토큰이 패턴에 매칭되면
      continue
    if len(w)< 2: # 단어토큰이 1글자라면
      continue
    temp.append(w) # 위 두개의 조건에 해당되지 않는 데이터만 누적
  clean_morphs.append(temp)
```
