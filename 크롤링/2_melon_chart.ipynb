```python
import requests # 전체 소스코드 가져옴 
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.melon.com/chart/index.htm"
#200번대 : 성공
#400번대 : 프론트엔드(클라이언트) 측 잘못 요청
#500번대 : 백엔드(서버)

#404: not found(URl 찾을 수 없음)
#405: 페이지는 찾았으나 알맞은 요청이 아닌 경우!
#     post로 요청해야하는데 get
#     [method not allowed]
#406 : 통신실패

# HTTP 프로토콜
# 1. 일반 헤더
# 2. 요청/응답 헤더
# 3. 엔터티 헤더
# 4. 바디

# 홈페이지 들어가서 요청한것처럼..
user_agent = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# 웹브라우저로 접속한 척 headers 추가하기!
res = requests.get(url, headers = user_agent)
print(res)
```

```python
soup = bs(res.text, 'lxml')
soup
```

```python
title_list = []
```

```python
title_list = soup.select("div.ellipsis.rank01 > span > a ")

title_100 = []
for i, title in enumerate(title_list) :
    print(f"{i+1}.{title.text}")
    title_100.append(title.text)
    
```

```python
print(len(title_100))
```

```python
singer_list = soup.select("div.ellipsis.rank02 > span")

singer_100 = []

for i, singer in enumerate(singer_list):
    print(f"{i+1}. {singer.text}")
    singer_100.append(singer)
```

```python
print(len(singer_list))
```

```python
# LIST(제목)
# LIST(가수)

# -> 표 형태로 변형(DataFrame)

import pandas as pd
dict = {
    "노래": title_100,
    "가수": singer_100
}
melon = pd.DataFrame(dict)
melon
```

```python
melon.to_csv("melon_top_100.csv", encoding="utf-8")
```
